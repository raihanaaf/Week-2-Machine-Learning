```python
import pandas as pd
import numpy as nd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('FinancialMarket.csv')
```

Split data menjadi fitur (x)  dan target menjadi (y)

```python
# Extract the x and combined data columns
x = data ['x'].values.reshape(-1,1)
y = data ['combined_data'].values
```

Split data menjadi training set dan testing set

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

Regresi linear

```python
model_linear = LinearRegression()
model_linear.fit(X_train, y_train)
y_pred_linear = model_linear.predict(X_test)
```

Regresi linear dengan basis fungsi

```python
poly_features = PolynomialFeatures(degree=2)
X_train_poly = poly_features.fit_transform(X_train)
X_test_poly = poly_features.transform(X_test)

model_poly = LinearRegression()
model_poly.fit(X_train_poly, y_train)
y_pred_poly = model_poly.predict(X_test_poly)
```

Evaluasi model

```python
import numpy as np

mse_linear = mean_squared_error(y_test, y_pred_linear)
rmse_linear = np.sqrt(mse_linear)
r2_linear = r2_score(y_test, y_pred_linear)

mse_poly = mean_squared_error(y_test, y_pred_poly)
rmse_poly = np.sqrt(mse_poly)
r2_poly = r2_score(y_test, y_pred_poly)
```

```python
print("Regresi Linear:")
print("MSE:", mse_linear)
print("RMSE:", rmse_linear)
print("R-squared:", r2_linear)
```

```python
print("\nRegresi Linear dengan Basis Fungsi:")
print("MSE:", mse_poly)
print("RMSE:", rmse_poly)
print("R-squared:", r2_poly)
```

Hasil Prediksi

```python
plt.scatter(X_test, y_test, label='Actual', color='skyblue')
plt.plot(X_test, y_pred_linear, label='Linear Regression', color='black')
plt.plot(X_test, y_pred_poly, label='Regression Linear dengan Basis Fungsi', color='pink')
plt.legend()
plt.show()
```
